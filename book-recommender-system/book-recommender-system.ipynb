## Hệ thống gợi ý dựa trên mức độ phổ biến

```python
ratings_with_name = ratings.merge(books,on='Book-Id')
```

```python
# Dem luot danh gia cua tung loai sach
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(numeric_only =True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df.head()
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
# Lay ra danh sach cac sach co so luong danh gia lon hon hoac bang 10 va sap xep lai theo trung binh
popular_df = popular_df[popular_df['num_ratings']>=10].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
# popular ket hop voi book , loai bo trung lap
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_ratings','avg_rating']]
```

## Hệ thống đề xuất dựa trên Collaborative Filtering

```python
#  luot danh gia sach >6
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 6
liked_users = x[x].index
```

```python
# loc luot danh gia user-id co nam trong liked_users
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(liked_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=2
famous_books = y[y].index
```

```python
# danh gia cuoi theo sach co nam trong famous_books
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
# tao luot danh gia 2 chieu
pt = final_ratings.pivot_table(index='Book-Id',columns='User-ID',values='Book-Rating')
```

```python
# cho nao ko dc danh gia (NAN) la 0
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
!pip install scikit-learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_id):
    # Chuyển ObjectId sang chuỗi nếu cần
    if not isinstance(book_id, str):
        book_id = str(book_id)
    
    # Kiểm tra xem book_id có trong chỉ mục hay không
    if book_id in pt.index:
        # Tìm chỉ mục của book_id
        index = np.where(pt.index == book_id)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:5]
        
        data = []
        for i in similar_items:
            item = []
            temp_df = books[books['Book-Id'] == pt.index[i[0]]]
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Year-Of-Publication'].values))
            
            data.append(item)
        
        return data
    else:
        print("Book not found")
        return None
```
